```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# check list of functionality in seaborn
print(dir(sns))
```

```python
ft= sns.load_dataset("flights")
ft
```

```python
# Bar Plot(Average passengers by month)
sns.barplot(data= ft, x= "month", y= "passengers", palette="plasma")
plt.xticks(rotation= 45)
plt.title("Average passenger by month")
plt.show()
```

```python
# box plot
sns.boxplot(data= ft, x= "month", y= "passengers", palette="plasma")
plt.xticks(rotation= 45)
plt.title("Average passenger by month")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.violinplot(
    data=ft, x="month", y="passengers", hue="month", palette="plasma", legend=False
)

# Rotate x-axis labels so they don't overlap
plt.xticks(rotation=45)

# Add clean, descriptive titles and labels
plt.title("Month-wise Passenger Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Passengers", fontsize=12)

# Automatically adjust layout to make sure everything fits perfectly
# plt.tight_layout()

# Display the plot
plt.show()
```

```python
sns.scatterplot(data= ft, x= "year", y= "passengers")
plt.title("scatterplot")
plt.show()
```

```python
sns.histplot(ft["passengers"], bins= 10, kde= True)
plt.title("Distribution")
plt.show()
```

```python
sns.kdeplot(ft["passengers"], fill= True)
plt.title("Density Plot")
plt.show()
```

```python
pivot= ft.pivot(index= "month", columns= "year", values= "passengers")
sns.heatmap(pivot, cmap= "YlGnBu", annot= True, fmt= "d")
plt.show()
```

```python
# pair plot since the dataset has only one numeric feature(passengers),
# the pair plot is simple to understand
sns.pairplot(ft)
plt.show()
```

```python
sns.countplot(data= ft, x= "month", palette= "plasma")

plt.xticks(rotation= 45)
plt.show()
```

```python
plt.Figure(figsize= (12, 8))


plt.subplot(2,2,1)

# line plot
sns.lineplot(data= ft, x= "year", y= "passengers")
plt.show()
```

```python
# line plot with Hue

sns.lineplot(data= ft, x="year", y="passengers", hue= "month")

plt.title("Monthly passenger trend")
plt.show()
```

```python
# Relational plot: "year", "passenger"

month_order= ["jan", "feb", "mar", "apr", 
              "may", "june", "july", "aug", "sep", "oct", "nov", "dec"]

sns.relplot(data= ft, x= "year", y= "passengers",
            kind= "line", hue= "month", legend="auto")

plt.show()        

# hue_order= month_order
```
