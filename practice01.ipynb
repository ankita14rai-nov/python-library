```python
df_cleaned = df.drop_duplicates()
print("shape of the dataframe after dropping duplicates:", df_cleaned.shape)
```

```python
df_clean= df_cleaned[df_cleaned['Quantity'] > 0]
```

```python
df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"])
```

```python
df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"], df_clean["UnitPrice"])
```

```python
df_clean["TotalPrice"]
```

```python
print("Step 3: Data Cleaning")
print("Original rows in the dataset:", df.shape[0] if df is not None else 0)
print("cancelled orders removed:", int(df["iscancelled"].sum()) if df is not None and "iscancelled" in df.columns else 0)
print("rows after cleaning:", df_clean.shape[0] if 'df_clean' in locals() else 0)
print(df_clean[["InvoiceNo", "Description", "Quantity", "UnitPrice", "CustomerID", "InvoiceDate", "TotalPrice"]].head(10) if 'df_clean' in locals() else "No cleaned dataset available")
print("\n")
```

```python
#Extract the year, month, and day from the InvoiceDate column and store them in new columns in the dataframe 

df_clean["year"]= df_clean["InvoiceDate"].dt.year

df_clean["month"]= df_clean["InvoiceDate"].dt.month

df_clean["day"]= df_clean["InvoiceDate"].dt.day


# Extra
df_clean["hour"] = df_clean["InvoiceDate"].dt.hour

# extract the name of the day of the week.
df_clean["day_of_week"] = df_clean["InvoiceDate"].dt.day_name()
```

```python
print("step 4: Data feature added")
print(df_clean[["InvoiceNo", "year", "month", "day_of_week", "hour"]].head(10) )
print("\n")
```

```python
import numpy as np
import pandas as pd
from pathlib import Path

# STEP 5: Compute business metrics

# NUNIQUE: Count the number of unique customers, orders, and products
# if an invoice has multiple products, it will be counted as one order
total_revenue = np.sum(df_clean['TotalPrice'])
total_orders = df_clean['InvoiceNo'].nunique()
total_customers = df_clean['CustomerID'].nunique()
avg_order_value= df_clean.groupby('InvoiceNo')['TotalPrice'].sum().mean()

print('total_revenue:', total_revenue)
print('total_orders:', total_orders)
print('total_customers:', total_customers)
print('average_order_value:', avg_order_value)
```

STEP 6: TOP Products (by revenue and by quantity )

```python
# find the top 10 products that generated the highest total revenue
top_products_revenue = (df_clean.groupby("Description")['TotalPrice'].sum().sort_values(ascending=False).head(10))


# find the top 10 products that were sold in the highest quantity
top_products_quantity = (df_clean.groupby("Description")['Quantity'].sum().sort_values(ascending=False).head(10))
```

```python
# Display the results

print("step 6: Top 10 products by revenue")
print(top_products_revenue)
print("\nTop 10 products by quantity sold")
print(top_products_quantity)
print("\n")
```

Step 7: Country-Wise Analysis

```python
# Calculate the total revenue generated by each country and sort it from highest to lowest.

country_revenue = (df_clean.groupby("Country")['TotalPrice']
                   .sum().astype(int).sort_values(ascending=False)) 
```

```python
# Display the results
print("step 7: Total revenue by country")
print(country_revenue.head(10))  # Display only the top 10 countries by revenue
```

STEP 8: Monthly sales trend

```python
# Create a new column containing only the year and month from the InvoiceDate column. This will help us analyze the sales trends over time on a monthly basis.
df_clean["yearMonth"] = df_clean["InvoiceDate"].dt.to_period("M")

# Calculate the total sales(revenue) for each year-Month
monthly_sales = df_clean.groupby("yearMonth")['TotalPrice'].sum()

print("step 8: Monthly sales trends")
print(monthly_sales)  # Display only the first 10 months of sales trends   
```

```python
df_clean["weekMonth"] = df_clean["InvoiceDate"].dt.to_period("W")
print("step 10: Sales by week")
print(df_clean.groupby("weekMonth")['TotalPrice'].sum())
```

```python
df_clean["Dayweek"] = df_clean["InvoiceDate"].dt.to_period("D")
print("step 11: Sales by day")
print(df_clean.groupby("Dayweek")['TotalPrice'].sum())
```

Step 9: Customer Anlaysis (RFM-Style basics)

```python
# create a customer summary by calculating spending, number of orders, average order value, and last purchase date for each customer. This will help us understand the purchasing behavior of our customers.
customer_summary = (
    df_clean.groupby("CustomerID")
    .agg(
        totalSpent=("TotalPrice", "sum"),
        NumOrders=("InvoiceNo", "nunique"),
        AvgOrderValue=("TotalPrice", "mean"),
        LastPurchase=("InvoiceDate", "max"),
    )
    .sort_values("totalSpent", ascending=False)
)

customer_summary.head()
```

STEP 10: Numpy-based statistical analysis on order values


```python
# calculate the total value of each order (invoice) and store the results as a numpy array.

order_values= df_clean.groupby("InvoiceNo")["TotalPrice"].sum().values

order_values
```

```python
# Step 10: calculate and display basic statistics for orders values.
print("step 10: order value statistics (Numpy)")

# Display the average order value
print("Mean:", np.mean(order_values))

# Display the middle (median) order value
print("Median:", np.median(order_values))

# Display the spread of order values
print("Standard Deviation:", np.std(order_values))

# Display the smallest order value
print("Minimum:", np.min(order_values))

# Display the largest order value
print("Maximum:", np.max(order_values))


# Display the range of order values
print("Range:", np.ptp(order_values))
```
