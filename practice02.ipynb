```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [24, 46, 13, 38]

# Maximum value find 
max_value = max(values)

# assign colors 
colors= ["green" if value != max_value else "purple" for value in values]

plt.bar(categories, values, color=colors, width= 0.8, bottom= 0)
plt.title("Bar chart: Compare categories", fontdict={
    "fontsize": 16, "fontweight": "bold", "color": "Red" , "family": "Arial"
})
plt.xlabel("Categories", fontdict={
    "fontsize": 16, 
    "fontweight": "bold", 
    "color": "Red",
    "family": "Arial"
})

plt.ylabel("Values", fontdict={
    "fontsize": 16, 
    "fontweight": "bold",
    "color": "Red",
    "family": "Arial"
})

plt.show()
```
