```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [24, 46, 13, 38]

# Maximum value find 
max_value = max(values)

# assign colors 
colors= ["green" if value != max_value else "purple" for value in values]

plt.bar(categories, values, color=colors, width= 0.8, bottom= 0)
plt.title("Bar chart: Compare categories", fontdict={
    "fontsize": 16, "fontweight": "bold", "color": "Red" , "family": "Arial"
})
plt.xlabel("Categories", fontdict={
    "fontsize": 16, 
    "fontweight": "bold", 
    "color": "Red",
    "family": "Arial"
})

plt.ylabel("Values", fontdict={
    "fontsize": 16, 
    "fontweight": "bold",
    "color": "Red",
    "family": "Arial"
})

plt.show()
```

```python
import matplotlib.pyplot as plt
domain= ["Data Science", "Machine Learning", "Deep Learning", "Computer Vision"]
values= [89, 78, 65, 90]

# Maximum value find 
max_value = max(values)

# # assign colors 
colors= ["green" if value != max_value else "purple" for value in values]

plt.plot(domain, values, color="blue", marker= "*", markersize= 10, markerfacecolor= "white", linewidth= 0.8, linestyle= "--")

plt.title("Line chart: Compare domains", fontdict={
    "fontsize": 16, "fontweight": "bold", "color": "Purple" , "family": "Arial"
})

plt.xlabel("Domains", fontdict={
    "fontsize": 16, 
    "color": "purple",
    "family": "Arial"
})

plt.ylabel("Values", fontdict={
    "fontsize": 16, 
    "color": "purple",
    "family": "Arial"
})
plt.show()
```

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt


# create a sequence of 30 consecutive data starting from 2026-01-01
days= pd.date_range(start="2026-01-01", periods=30)

# generate 30 random numbers from a normal distribution with mean= 0 and standard deviation= 1
# compute their cumulative sum and shift all values upward by 50.

values= np.cumsum(np.random.normal(0, 1, 30)) + 50

# plot a line chart
plt.plot(days, values, color= "blue", marker= "*", markersize= 6.7, markerfacecolor= "white", linewidth= 0.8, linestyle= "--")

# add a title and axis labels
plt.title("Line chart: Trend over time")
plt.xlabel("Date")
plt.ylabel("Value")

# Rotate the x-axis labels by 45 degrees for better readability
plt.xticks(rotation= 45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate normal distribution data
data = np.random.normal(loc=60, scale=10, size=1000)

# 2. Create the histogram
plt.hist(data, bins=10, edgecolor="red")

# 3. Add labels and title
plt.title("Histogram: Data distribution")
plt.xlabel("values")
plt.ylabel("frequency")

# 4. Display the plot
plt.show()
```
